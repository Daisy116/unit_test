```python
# 抓live首頁
from selenium import webdriver
from bs4 import BeautifulSoup
from openpyxl import Workbook

# 創建一個空白活頁簿物件
wb = Workbook()
# 選取正在工作中的表單
ws = wb.active
ws.append(['賽事名稱', 'KO時間', '狀態', '主隊', '比分', '客隊', '半場'])

browser = webdriver.Chrome('./chromedriver')
browser.get('http://live2.gooooal.com/')
soup = BeautifulSoup(browser.page_source)

# 賽事名稱
i = 2
for ele in soup.select('.name-td a'):
    ws.cell(row = i, column = 1, value = ele.text)
    i = i + 1
# KO時間
i = 2
j = 2
for ele in soup.select('.time-td span'):
    if ((i % 2) != 0) :
        ws.cell(row = j, column = 2, value = ele.text)
        j = j + 1
    i = i + 1
# 狀態
i = 2
j = 2
for ele in soup.select('.state-td span'):
    if ((i % 2) != 0) :
        ws.cell(row = j, column = 3, value = ele.text)
        j = j + 1
    i = i + 1
# 主隊   
i = 2
for ele in soup.select('.home-team-td .team'):
    ws.cell(row = i, column = 4, value = ele.text)
    i = i + 1
# 比分
i = 2
for ele in soup.select('.score-td a'):
    ws.cell(row = i, column = 5, value = ele.text)
    i = i + 1
# 客隊   
i = 2
for ele in soup.select('.visiting-team-td .team'):
    ws.cell(row = i, column = 6, value = ele.text)
    i = i + 1
# 半場
i = 2
for ele in soup.select('.half-td'):
    ws.cell(row = i, column = 7, value = ele.text)
    i = i + 1
browser.close()
wb.save('live2_index3.xlsx')
```

```python
# 從這兒改live + odd
from selenium import webdriver
from bs4 import BeautifulSoup
from openpyxl import Workbook

# # 創建一個空白活頁簿物件
# wb = Workbook()
# # 選取正在工作中的表單
# ws = wb.active
# ws.title = "live"
# ws.append(['公司名稱', '主隊水位', '客隊水位'])

browser = webdriver.Chrome('./chromedriver')
browser.get('http://live2.gooooal.com/')
soup = BeautifulSoup(browser.page_source)

link_list1 = []
link_list2 = []
link_list3 = []
for ele in soup.select('.score-td a'):
    href = ele['href']
#     print(str[22:29])
    link_list1.append('http://odds.gooooal.com/fb_match.html?mid=' + href[22:29] + '&type=1')
    link_list2.append('http://odds.gooooal.com/fb_match.html?mid=' + href[22:29] + '&type=2')
    link_list3.append('http://odds.gooooal.com/fb_match.html?mid=' + href[22:29] + '&type=3')
    
browser.close()


for i in range(len(link_list1)):  
    browser = webdriver.Chrome('./chromedriver')
    browser.get(link_list1[i])
    soup = BeautifulSoup(browser.page_source)

    
    for e2 in soup.select('.font24'):
        print(e2.text)
    
    sheet_page = str(i)
    ws1 = wb.create_sheet("odd" + sheet_page) 
    ws1.append(['公司名稱', '主隊水位', '客隊水位'])
    i = 2
    for e2 in soup.select('.tdLine1+ .tdLine1 a'):
    #     print(e2.text)
        ws1.cell(row = i, column = 1, value = e2.text)
        i = i + 1

    i = 2
    for e2 in soup.select('.li01:nth-child(1) .oddsPopHover'):
    #     print(e2.text)
        ws1.cell(row = i, column = 2, value = e2.text)
        i = i + 1

    i = 2
    for e2 in soup.select('.li02+ .li01 .oddsPopHover'):
    #     print(e2.text)
        ws1.cell(row = i, column = 3, value = e2.text)
        i = i + 1
    browser.close()
wb.save('odd3.xlsx')
```

```python
# odd.py
from selenium import webdriver
from bs4 import BeautifulSoup
from openpyxl import Workbook

browser = webdriver.Chrome('./chromedriver')
browser.get('http://odds.gooooal.com/fb_match.html?mid=1533684&type=1')
soup = BeautifulSoup(browser.page_source)

wb = Workbook()
# sheet_page = str(i)
ws = wb.active
ws.append(['公司名稱', '主隊水位', '客隊水位'])
for e2 in soup.select('.oddsLogLeft .font24'):
    title = e2.text
for e2 in soup.select('.oddsLogRight .font24'):
    title = title + 'vs' + e2.text
ws.title = title

sp = soup.select('tr[id^=tr_asia_] .tdLine1+ .tdLine1 a')
sp2 = soup.select('tr[id^=tr_asia_] .li01:nth-child(1) .oddsPopHover')
# sp2 = soup.select('td[id^=td_asian_pop_]')
sp3 = soup.select('tr[id^=tr_asia_] .li02+ .li01 .oddsPopHover')


sp3 = soup.select('tr[id^=tr_eour_1_] #div_tab_europe a')

i = 2
for s in sp:
    if s.find('span'):
        s.find('span').extract()
#     print(s.text)
    ws.cell(row = i, column = 1, value = s.text)
    i = i + 1

i = 2
for s in sp2:
    print(s.text)
    ws.cell(row = i, column = 2, value = s.text)
    i = i + 1

i = 2
for s in sp3:
#     print(s.text)
    ws.cell(row = i, column = 3, value = s.text)
    i = i + 1
    
browser.close()
wb.save('odd_ok.xlsx')
```
