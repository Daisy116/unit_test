```python
from openpyxl import Workbook

# 創建一個空白活頁簿物件
wb = Workbook()

# 選取正在工作中的表單
ws = wb.active

# 指定值給 A1 儲存格
ws['A1'] = '我是儲存格'

# 向下新增一列並連續插入值
ws.append(['是是是', '否否否', '宜宜宜'])
ws.append([3, 2, 1])

ws.cell(row=4, column=2, value=10)

# 儲存成 create_sample.xlsx 檔案
wb.save('create2.xlsx')
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from openpyxl import Workbook

# 創建一個空白活頁簿物件
wb = Workbook()
# 選取正在工作中的表單
ws = wb.active

# browser = webdriver.Chrome()
browser = webdriver.Chrome('./chromedriver')
browser.get('http://app.gooooal.com/footBallDataBaseAction.do?method=doFootBallDataBase')
soup = BeautifulSoup(browser.page_source)

for ele in soup.select('.right div div span'):
    print(ele.text)
#     ws.append([ele.text])
    ws.cell(row=1, column = 2, value = ele.text)

i = 2
for ele in soup.select('.right div table tbody tr td center span a'):
    print(ele.text)
#     ws.append([ele.text])
    ws.cell(row = i, column = 2, value = ele.text)
    i = i + 1
    
browser.close()

browser = webdriver.Chrome('./chromedriver')
browser.get('http://app.gooooal.com/footBallDataBaseAction.do?method=doFootBallDataBase')
soup = BeautifulSoup(browser.page_source)

for ele in soup.select('.left div div span'):
    print(ele.text)
    ws.cell(row=1, column = 1, value = ele.text)

j = 2
for ele in soup.select('.left div table tbody tr td center span a'):
    print(ele.text)
    ws.cell(row=j, column = 1, value = ele.text)
    j = j + 1
    
browser.close()

# 儲存成 create_sample.xlsx 檔案
wb.save('create.xlsx')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
 
browser = webdriver.Chrome()
try:
    browser.get('https://www.baidu.com')
    input = browser.find_element_by_id('kw')
    input.send_keys('Python')
    input.send_keys(Keys.ENTER)
    wait = WebDriverWait(browser, 10)
    wait.until(EC.presence_of_element_located((By.ID, 'content_left')))
    
    # 輸出當前的URL
    print(browser.current_url)
    
    # 輸出當前的Cookies
    print(browser.get_cookies())
    
    # 網頁源代碼
    print(browser.page_source)
finally:
    browser.close()
```
